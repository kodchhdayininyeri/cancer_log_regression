```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
data=pd.read_csv("cancert.csv")
```

```python
veri=data.copy()
```

```python
veri.head()
```

```python
veri=veri.drop(columns=["id","Unnamed: 32"])
```

```python
veri.head()
```

```python
veri.isnull().sum()
```

```python
y=veri["diagnosis"]
x=veri.drop(columns="diagnosis",axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,)
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
model=LogisticRegression(random_state=0)
model.fit(x_train,y_train)
tahmin=model.predict(x_test)
```

```python
tahmin
```

```python
y_test_to_np=y_test.to_numpy()
y_test_to_np
```

```python
tahmin_df=pd.DataFrame({"Tahmin": tahmin})
y_test_df = pd.DataFrame({"Y_Test": y_test_to_np})
df = pd.concat([tahmin_df, y_test_df], axis=1)
```
